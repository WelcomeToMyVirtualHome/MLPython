```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 4, 3, 5]
y = [1, 3, 3, 2, 5]
plt.plot(x,y,marker='o',linewidth=0)
```

```python
B0 = 0
B1 = 0
alpha = 0.005
epochs = np.arange(1,20);
error = []
for epoch in epochs:
    for x_i,y_i in zip(x,y):
        p = B0 + B1 * x_i
        err = p - y_i
        error.append(err)
        B0 = B0 - alpha * err
        B1 = B1 - alpha * err * x_i
        
plt.plot(error)
plt.grid(True)
print("B0=", B0, " B1=", B1)
```

```python
Y = np.add(np.multiply(x,B1),B0)
plt.plot(x,y,'o',x, Y,'*-')
```

```python
RMSE = np.sqrt(np.sum((Y-y)**2)/np.size(y))
print(RMSE)
```
